```python
import os
from dotenv import load_dotenv
import requests
```

```python
load_dotenv("../.env")
```

```python
SR_APIKEY = os.environ.get('SR_API_KEY')
SR_APISECRET = os.environ.get('SR_API_SECRET')
SCHEMA_REGISTRY_URL = os.environ.get('SCHEMA_REGISTRY_URL')
SUBJECT_NAME = 'system.usage.new-value' 

url = f"{SCHEMA_REGISTRY_URL}/subjects/{SUBJECT_NAME}"

# Define the query parameters
params = {
    'permanent': 'true'
}
```

```python
auth = (SR_APIKEY, SR_APISECRET)
```

Required : First soft delete, and then hard delete

```python
response = requests.delete(url, auth=auth)
response = requests.delete(url, params=params, auth=auth)
```
